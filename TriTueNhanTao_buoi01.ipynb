```python
#bài 1
def cong(a,b):
  return a+b
def tru(a,b):
  return a-b
def nhan(a,b):
  return a*b
def chia(a,b):
  return a/b
a=int (input("Nhập số nguyên thứ nhất: a = "))
b=int (input("Nhập số nguyên thứ nhất: b = "))
tong = cong(a,b)
hieu = tru(a,b)
tich = nhan(a,b)
thuong = chia(a,b)
print ("Tổng hai số là: ", tong)
print ("Hiệu hai số là: ", hieu)
print ("Tích hai số là: ", tich)
print ("Thương hai số là: ", thuong)
```
